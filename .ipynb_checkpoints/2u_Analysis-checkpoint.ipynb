```python
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
timeframe = "1D"

start_date = pd.Timestamp("2014-09-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-02-19", tz="America/New_York").isoformat()

ticker = ['TWOU']

Twou_df = alpaca.get_barset(
    ticker,
    timeframe,
    start=start_date,
    end=end_date

).df
Twou_df.head()
```

```python
csvpath = Path ( "TWOU.csv")
```

```python
TWOU_df = pd.read_csv(csvpath)
TWOU_df.head()
```

```python
TWOU_df.set_index(pd.to_datetime(TWOU_df['Date'], infer_datetime_format = True), inplace = True)
TWOU_df.columns
TWOU_columns = ["Date", "TWOU"]
TWOU_df.drop(['Date'], axis = 1, inplace = True)
TWOU_df.index = TWOU_df.index.date
TWOU_df.head()
```

```python
closing_price = pd.DataFrame()
for ticker in ticker:
    closing_price[ticker] =TWOU_df['Close']
    
closing_price
```

```python
daily_returns = closing_price.pct_change().dropna()
daily_returns
```

```python
daily_returns.plot(figsize=(20,9), title = "Daily Returns")
```

```python
cum_return = (1+ daily_returns).cumprod()
cum_return
```

```python
cum_return.plot(figsize=(20,9), title = "Cumulative Returns")
```

```python
TwoU_std = daily_returns.std()
TwoU_std
```

```python
TwoU_annual_std = daily_returns.std() * np.sqrt(252)
TwoU_annual_std
```

```python
daily_returns.rolling(21).std().plot(figsize=(20,10), title= "21 Day Rolling Statistics")
```

```python
Twou_sharpe_ratio = (daily_returns.mean() * 252) / TwoU_annual_std
Twou_sharpe_ratio
```

```python
Twou_sharpe_ratio.plot.bar(title = "Sharpe Ratio")
```
